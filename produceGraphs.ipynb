```python
import pandas as pd
import matplotlib.pyplot as plt

my_file = pd.read_csv("Data/gapminder.txt",delimiter = "\t")
Canada = my_file.loc[my_file["country"]=="Canada"]

Canada.plot.line(x = "year", y = "lifeExp",label = "Life Expectancy of Canada", figsize =(8,6))
plt.suptitle("Life Expectancy of Canada over the years",fontsize=20)
plt.xlabel("Year", fontsize = 16)
plt.ylabel("YeaLife Expectancy", fontsize = 16)
plt.show()
```

```python
#can use the dot operator to specificy a column
Africa   = my_file[my_file.continent == "Africa"]
Americas = my_file[my_file.continent == "Americas"]
Asia     = my_file[my_file.continent == "Asia"]
Europe   = my_file[my_file.continent == "Europe"]
Oceania  = my_file[my_file.continent == "Oceania"]

#calculate the mean gdp (6th column)
Africa_meanGDP = Africa.iloc[:,5].mean()
Americas_meanGDP = Americas.iloc[:,5].mean()
Asia_meanGDP = Asia.iloc[:,5].mean()
Europe_meanGDP = Europe.iloc[:,5].mean()
Oceania_meanGDP = Oceania.iloc[:,5].mean()
```

```python
#create a bar graph
continents = ["Africa","Americas","Asia","Europe","Oceania"]
meanGDP = [Africa_meanGDP,Americas_meanGDP, Asia_meanGDP, Europe_meanGDP, Oceania_meanGDP]

#size of the figure
#fig_size = plt.rcParam["figure.figsize"]
#fig_size[0] = 10
#fig_size[1] = 8

plt.bar(continents, meanGDP, align = "center")
plt.ylabel("Mean GDP/Capita")
plt.title("Mean GDP/Capita for Continents")
plt.savefig("MeanGDP_Plot.png")
```
