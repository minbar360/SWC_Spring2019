```python
my_file = pd.read_csv("Data/gapminder.txt",delimiter = "\t")
#my_file = pd.read_csv("path to the csv file")
```

```python
print(my_file)
```

```python
my_file.head() #shows the first five lines of the file
```

```python
#General Syntax
#my_file.iloc[rows,columns]
#recall indexing starts at 0
my_file.iloc[0,2]
```

```python
#first 10 rows, the 3rd and 4th column
#the last value is not included
my_file.iloc[0:10,2:4]
```

```python
#displays row 0, row 2, row 5, column 2 (year) and column 4 (pop)
my_file.iloc[[0,2,5],[2,4]]
```

```python
#General Syntax
#my_file.loc[row,column]
#the colon before the comma means display all rows
my_file.loc[:,['year','pop']]
```

```python
my_file.loc[1,:]
```

```python
#shows only rows where the value in the country column is Afghanistan
my_file.loc[my_file["country"]=="Afghanistan"]
```

```python
my_file.loc[my_file["country"]=="Afghanistan",["year","pop"]]
```

```python
#check to see if the value in the contry column is Afghanistan
check=my_file["country"]=="Afghanistan"
#display the first 13 lines in the check
print(check.head(13))
#display the last 13 lines in the check
print(check.tail(13))
```

```python
#Extract the 3rd and 5th columns of the dataset
my_file.iloc[:,[2,4]]
```

```python
#Extract all countries for which life expectancy is below 70
my_file.loc[my_file["lifeExp"]<70,:]
```
